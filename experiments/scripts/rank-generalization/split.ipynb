### Imports

```python
import os

import pandas as pd
import torch
from tqdm import tqdm
```

### Do splitting

```python
files = []
files.append("weight-diff-20250514-21-scaling-qwen-4b-rank-2")
files.append("weight-diff-20250514-21-scaling-qwen-4b-rank-4")
files.append("weight-diff-20250514-21-scaling-qwen-4b-rank-8")
files.append("weight-diff-20250514-21-scaling-qwen-4b-rank-16")
files.append("weight-diff-20250514-21-scaling-qwen-4b-rank-32")
files.append("weight-diff-20250514-21-scaling-qwen-4b-rank-64")
files.append("weight-diff-20250514-23-scaling-gemma-4b-rank-2")
files.append("weight-diff-20250515-01-scaling-gemma-4b-rank-4")
files.append("weight-diff-20250515-01-scaling-gemma-4b-rank-8")
files.append("weight-diff-20250515-01-scaling-gemma-4b-rank-16")
files.append("weight-diff-20250515-01-scaling-gemma-4b-rank-32")
files.append("weight-diff-20250515-01-scaling-gemma-4b-rank-64")
files.append("weight-diff-20250515-01-scaling-gemma-4b-rank-128")
```

```python
for f in files:
    dir = os.path.join("/workspace/datasets", f)
    new_dir = os.path.join("/workspace/datasets", f + "_split")
    os.makedirs(new_dir, exist_ok=True)
    out_cnt = 0
    for file in os.listdir(dir):
        if not file.endswith(".pt"):
            continue
        file = os.path.join(dir, file)
        print(file)
        data = torch.load(file, map_location="cpu")
        for i, d in enumerate(tqdm(data)):
            out_cnt += 1
            for k, v in d["weight_diff"].items():
                d["weight_diff"][k] = (v[0].detach().clone(), v[1].detach().clone())
            torch.save([d], os.path.join(new_dir, f"weight_diff_{out_cnt}.pt"))
```

```python
df = pd.read_csv("/workspace/datasets/news-stories-v0.1-87-shards-split.csv")
```

```python
val = df[df.split == "test"].topic.unique()
```

```python
dir = "/workspace/datasets/weight-diff-20250514-news-qwen-4b"
out_dir = f"{dir}-val"

os.makedirs(out_dir, exist_ok=True)

gradient_files = []
for root, _, files in os.walk(dir):
    for f in files:
        if f.startswith("weight_diff_") and f.endswith(".pt"):
            gradient_files.append(os.path.join(root, f))

gradient_files.sort()
print(f"Found {len(gradient_files)} gradient files")

n_written = 0
for file_path in tqdm(gradient_files, desc="Loading files"):
    for item in torch.load(file_path, map_location="cpu"):
        if item["topic"] in val:
            n_written += 1
            print(n_written, item["topic"])
            for k, v in item["weight_diff"].items():
                item["weight_diff"][k] = (v[0].detach().clone(), v[1].detach().clone())
            torch.save([item], os.path.join(out_dir, f"weight_diff_{n_written}.pt"))
```
